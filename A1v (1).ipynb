```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')

df.tail(10)
```

## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
filtered_df = df[(df['Date'] >= "2023-00-00")&(df['Fuel Type']=='Regular Unleaded Gasoline')]
columns=['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's"]
# print(filtered_df)

mean_data=filtered_df[columns].mean(axis=0)
mean_data
# Create a DataFrame using numpy.column_stack()
array1 = np.array(columns)
array2 = np.array(mean_data)
mean_df = pd.DataFrame(np.column_stack((array1, array2)), columns=['Cities', 'Mean_Price'])
sorted_df=mean_df.sort_values(by='Mean_Price')

# Display the DataFrame
# print(sorted_df)

#create bar plot to visualize frequency of each team
plt.barh(sorted_df['Cities'],sorted_df['Mean_Price'].astype(float))
plt.title('Bar Plot for Categorical Data')
plt.xlabel('Cities')
plt.ylabel('Mean_Price')





plt.gcf().set_size_inches(25, 6)


plt.show()
```

## Question 2
Plot the corresponding dot plot for the same information presented in Question 1.

```python
plt.plot(sorted_df['Cities'],sorted_df['Mean_Price'], marker='o', linestyle='None', color='green', markersize=8)

# Customize the plot
plt.title('Dot Plot of Mean Prices in 2023')
plt.xlabel('Cities')
plt.ylabel('Mean Prices')

plt.gcf().set_size_inches(25, 6)
# Display the plot
plt.show()
```
